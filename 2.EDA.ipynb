### 1. The distribution of people who are potentially fraud

```python
# Create Some features for conveniece of calculations
inpatient = data[data['IsOutpatient'] == 0]
outpatient = data[data['IsOutpatient'] == 1]
diag_code = data.columns[data.columns.str.contains('DiagnosisCode')].tolist()
proc_code = data.columns[data.columns.str.contains('ProcedureCode')].tolist()
chronic = data.columns[data.columns.str.contains("Chronic")].tolist()
data["ClaimDuration"] = data["ClaimEndDt"] - data["ClaimStartDt"]
data["ClaimDuration"] = data["ClaimDuration"].dt.days
data["NoPhy"] = data[['AttendingPhysician','OperatingPhysician']].isna().all(axis =1)
data['AllPhy'] = data[['AttendingPhysician','OperatingPhysician']].notnull().all(axis =1)
data['SameAttOper'] = data['AttendingPhysician'] == data['OperatingPhysician']
data["AdmisDuration"] = data["DischargeDt"] - data["AdmissionDt"]
data["AdmisDuration"] = data["AdmisDuration"].dt.days
data["AgeAtClm"] = round((data["ClaimStartDt"] - data["DOB"]).dt.days/365,0).astype(int)
data["TotalRev"] = data['InscClaimAmtReimbursed'] + data['DeductibleAmtPaid']
data['ClmYear'] = data.ClaimStartDt.dt.year.rename('Year')
data['ClmMonth'] = data.ClaimStartDt.dt.month.rename('Month')
data['ClmWeek'] = data.ClaimStartDt.dt.week.rename('Week')
data['InsCovRatio'] = data['InscClaimAmtReimbursed']/(data['InscClaimAmtReimbursed'] + data["DeductibleAmtPaid"])
data['RevPerDay'] = data["TotalRev"]/(data['ClaimDuration']+1)
data['Chronic_Sum'] = data[chronic].sum(axis = 1)
```

```python
# Function for transforming dataframe, which will be used for plotting
def transform(agg_col, agg_func, new_colname):
    new_df = data.groupby(['PotentialFraud','IsOutpatient','Provider'])[[agg_col]].agg(agg_func)
    new_df.columns.values[0] = new_colname
    new_df = new_df.reset_index()
    return new_df
```

```python
# Function for creating barplots grouped by in/out patients and fraudulent providers
def create_barplot(df, column, title1, title2):
    fig, ax = plt.subplots(1, 2, figsize=(12, 6))
    sns.barplot(data = df[(df['IsOutpatient'] == 0)], 
                x = column, y = 'PotentialFraud', ax = ax[0], fliersize= 1);
    sns.barplot(data = df[(df['IsOutpatient'] == 1)], 
                x = column, y = 'PotentialFraud', ax = ax[1], fliersize= 1);
    ax[0].set_title(title1);
    ax[1].set_title(title2);
```

```python
# Plotting the Potential Fraud 
fraud = data['PotentialFraud'].value_counts()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(fraud.index, fraud.values, alpha=0.8)
plt.title('Potential Fraud Distribution', fontsize=20)
plt.xlabel('Potential Fraud', fontsize=12)
plt.ylabel('Number of Occurences', fontsize=12)
plt.show();
```

From the barplot above potential fraud occurs more than 200k times compared to cases with no potential fraud that have about 350k cases.

### 2. What age of people are most likely to be affected by Potential fraud

```python
patient_age = data.groupby(['PotentialFraud','AgeAtClm'])[['ClaimID']].count()
patient_age.columns.values[0] = 'Count'
patient_age = patient_age.reset_index()
sns.lineplot(data = patient_age,x = "AgeAtClm", y = 'Count', hue='PotentialFraud');
```

From the line chart above, the age of people most likely to be affected by fraud is 65. This is significant because this is the age for Medicare. According to the economist (May, 31st 2014 edition) Medicare doles out about 272 billion  USD, approximately 60 billion USD is lost annually.
